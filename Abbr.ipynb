```python
import spacy
```

```python


def find_acronyms(text):
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)

    acronyms = []

    for token in doc:
        if token.is_upper and len(token.text) > 1 and token.text.isalpha():
            acronyms.append(token.text)

    return acronyms

def trouver_acronymes(texte):
    nlp1 = spacy.load('fr_core_news_sm')
    doc = nlp1(texte)

    acronymes = []

    for token in doc:
        if token.is_upper and len(token.text) > 1 and token.text.isalpha():
            acronymes.append(token.text)

    return acronymes

```

```python
# Example usage
text = "The FBI is investigating the case. NASA is planning a mission to Mars."
result = find_acronyms(text)
print(result)
```

```python
# Exemple d'utilisation
texte = "Le FBI enquête sur une affaire de vol de données. L'OTAN discute de la sécurité internationale."
resultat = trouver_acronymes(texte)
print(resultat)
```

```python
import spacy
from scispacy.abbreviation import AbbreviationDetector

nlp = spacy.load("en_core_web_sm")
abbreviation_pipe = AbbreviationDetector(nlp)
nlp.add_pipe('abbreviation_detector')

text2 = """The Eiffel Tower (E.T.) is a popular tourist
        attraction in Paris (PR), France (FR). It was built in 1889.
        E.T. is one of the famous monuments in FR."""
text = "StackOverflow (SO) is a question and answer site for professional and enthusiast programmers. SO rocks!. The FBI is investigating the case. NASA is planning a mission to Mars."

def replace_acronyms(text):
    doc = nlp(text)
    altered_tok = [tok.text for tok in doc]
    for abrv in doc._.abbreviations:
        altered_tok[abrv.start] = str(abrv._.long_form)

    return " ".join(altered_tok)

replaced_text = replace_acronyms(text)
print(replaced_text)
```
