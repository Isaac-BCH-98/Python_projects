```python
import xml.etree.ElementTree as ET

def build_drawio_schema(mcf):
    # Create the root element for the XML tree
    root = ET.Element("diagram", {"name": "MCF Python", "xmlns": "http://graphml.graphdrawing.org/xmlns"})
    graph = ET.SubElement(root, "graph", {"id": "G", "edgedefault": "directed"})

    # Process entities
    for entity_name in mcf["entities"]:
        entity = ET.SubElement(graph, "node", {"id": entity_name, "labels": "Entity"})
        mx_cell = ET.SubElement(entity, "mxCell", {"vertex": "1", "parent": "G"})
        mx_geometry = ET.SubElement(mx_cell, "mxGeometry", {"x": "120", "y": "120", "width": "120", "height": "60", "as": "geometry"})

    # Process attributes
    for attribute_name in mcf["attributes"]:
        attribute = ET.SubElement(graph, "node", {"id": attribute_name, "labels": "Attribute"})
        mx_cell = ET.SubElement(attribute, "mxCell", {"vertex": "1", "parent": "G"})
        mx_geometry = ET.SubElement(mx_cell, "mxGeometry", {"x": "240", "y": "220", "width": "80", "height": "40", "as": "geometry"})

    # Process relationships
    for relation_name, relation in mcf["relationships"].items():
        relationship = ET.SubElement(graph, "edge", {"id": relation_name, "labels": "Relation", "source": relation["source"], "target": relation["target"]})
        mx_cell = ET.SubElement(relationship, "mxCell", {"edge": "1", "parent": "G"})
        mx_geometry = ET.SubElement(mx_cell, "mxGeometry", {"relative": "1", "as": "geometry"})

    # Process attribute-entity links
    for link_name, link in mcf["links"].items():
        link_element = ET.SubElement(graph, "edge", {"id": link_name, "labels": "Links", "source": link["attribute"], "target": link["entity"]})
        mx_cell = ET.SubElement(link_element, "mxCell", {"edge": "1", "parent": "G"})
        mx_geometry = ET.SubElement(mx_cell, "mxGeometry", {"relative": "1", "as": "geometry"})

    # Create the XML tree
    tree = ET.ElementTree(root)

    # Return the XML string
    return ET.tostring(root, encoding="utf-8", xml_declaration=True)

# Example Python MCF
mcf = {
    "entities": ["Entity1", "Entity2"],
    "attributes": ["Attribute1"],
    "relationships": {
        "Relation1": {"source": "Entity1", "target": "Entity2"}
    },
    "links": {
        "Link1": {"attribute": "Attribute1", "entity": "Entity1"}
    }
}

# Build the DrawIO schema in XML format
drawio_xml = build_drawio_schema(mcf)

# Print the XML schema
print(drawio_xml.decode("utf-8"))
```

```python
from IPython.display import display, HTML
def generate_drawio_xml(items, links, node_shape='rectangle', arrowed_lines=False):
    xml_template = '''<?xml version="1.0" encoding="UTF-8"?>
    <mxfile host="app.diagrams.net" modified="2023-07-01T00:00:00.000Z" agent="5.0 (Windows)" etag="Et4d0GurXuNpCmAB1d8z" version="12.9.5" type="device">
        <diagram name="Diagram" id="Et4d0GurXuNpCmAB1d8z">
            <mxGraphModel dx="805" dy="415" grid="1" gridSize="10" guides="1" tooltips="1" connect="1" arrows="1" fold="1" page="1" pageScale="1" pageWidth="827" pageHeight="1169" math="0" shadow="0">
                <root>
                    <mxCell id="0" />
                    <mxCell id="1" parent="0" />
                </root>
                <mxCell id="2" value="" style="group" vertex="1" connectable="0" parent="1">
                    <mxGeometry x="20" y="20" width="800" height="400" as="geometry" />
                </mxCell>
                {0}
                {1}
            </mxGraphModel>
        </diagram>
    </mxfile>
    '''

    node_template = '''
    <mxCell id="{0}" value="{1}" style="shape={2}" vertex="1" parent="2">
        <mxGeometry x="{3}" y="{4}" width="80" height="40" as="geometry" />
    </mxCell>
    '''

    link_template = '''
    <mxCell id="{0}" value="{1}" style="{2}" edge="1" parent="2" source="{3}" target="{4}">
        <mxGeometry relative="1" as="geometry" />
    </mxCell>
    '''

    nodes_xml = ""
    links_xml = ""
    node_id = 3  # Start node_id from 3 to avoid conflicts with default cells
    link_id = 1

    for item in items:
        node_xml = node_template.format(node_id, item, node_shape, (node_id * 100), (node_id * 100))
        nodes_xml += node_xml
        node_id += 1

    for link in links:
        source = link[0]
        target = link[1]
        label = link[2]
        link_xml = link_template.format(link_id, label, "endArrow=none" if not arrowed_lines else "", items.index(source) + 3, items.index(target) + 3)
        links_xml += link_xml
        link_id += 1

    drawio_xml = xml_template.format(nodes_xml, links_xml)
    return drawio_xml

def generate_html_code(items, links, node_shape='rectangle', arrow_heads=False, svg_width=800, svg_height=600):
    svg_width = 800  # Largeur de l'affichage SVG
    svg_height = 600  # Hauteur de l'affichage SVG

    html = f"<svg width='{svg_width}' height='{svg_height}'>"
    y = 50

    if node_shape == 'rectangle':
        shape_width = 80
        shape_height = 30
    else:
        shape_width = 60
        shape_height = 40

    for item in items:
        if node_shape == 'rectangle':
            html += f"<rect x='100' y='{y}' width='{shape_width}' height='{shape_height}' style='fill:white;stroke:black;stroke-width:2px;'/>"
        else:
            html += f"<ellipse cx='{100 + shape_width/2}' cy='{y + shape_height/2}' rx='{shape_width/2}' ry='{shape_height/2}' style='fill:white;stroke:black;stroke-width:2px;'/>"
        html += f"<text x='{100 + shape_width/2}' y='{y + shape_height/2}' dominant-baseline='middle' text-anchor='middle'>{item}</text>"
        y += 90

    for link in links:
        x1 = 100 + shape_width
        y1 = 50 + items.index(link[0]) * 90 + shape_height / 2
        x2 = 100
        y2 = 50 + items.index(link[1]) * 90 + shape_height / 2

        if arrow_heads:
            html += f"<marker id='arrow' markerWidth='10' markerHeight='10' refX='8' refY='3' orient='auto'><path d='M0,0 L0,6 L9,3 z' fill='black' /></marker>"
            html += f"<line x1='{x1}' y1='{y1}' x2='{x2}' y2='{y2}' style='stroke:black;stroke-width:2px;marker-end:url(#arrow);' />"
        else:
            html += f"<line x1='{x1}' y1='{y1}' x2='{x2}' y2='{y2}' style='stroke:black;stroke-width:2px;' />"

        text_x = (x1 + x2) / 2
        text_y = (y1 + y2) / 2
        html += f"<text x='{text_x}' y='{text_y}' dominant-baseline='middle' text-anchor='middle'>{link[2]}</text>"

    html += "</svg>"
    return html

def save_xml_to_file(xml_content, filename):
    with open(filename, 'w', encoding='utf-8') as file:
        file.write(xml_content)

# Exemple d'utilisation
items = ['A', 'B', 'C', 'D']
links = [('A', 'B', 'alpha'), ('A', 'C', 'beta'), ('C', 'D', 'gamma')]
node_shape = input("Choisissez la forme des nœuds ('rectangle' ou 'ellipse') : ")
arrow_heads = input("Les arcs sont-ils fléchés ? (Oui/Non) : ")

if arrow_heads.lower() == 'oui':
    arrow_heads = True
else:
    arrow_heads = False



# Modifier les valeurs de taille d'affichage ici
svg_width = 1000
svg_height = 800

drawio_xml = generate_drawio_xml(items, links, node_shape=node_shape, arrowed_lines=False)
# Sauvegarder le contenu XML dans un fichier .txt
save_xml_to_file(drawio_xml, 'diagram.txt')
print("Contenu XML sauvegardé avec succès dans le fichier diagram.xml.")
html_code = generate_html_code(items, links, node_shape=node_shape, arrow_heads=arrow_heads, svg_width=svg_width, svg_height=svg_height)
display(HTML(html_code))
```
